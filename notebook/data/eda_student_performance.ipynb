```python
#importing importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#Reading data from the source

df=pd.read_csv('C:/Users/Ayushi Singh/Documents/Study Material/mlproject/notebook/data/stud.csv')
```

```python
#Segregare categorical and numerical variable
cat_var=[col for col in df.columns if df[col].dtype in ['object']]
int_var=[col for col in df.columns if df[col].dtype in ['int64','int32','float32','flot64']]
```

```python
#check for null values

df.isnull().sum()
```

```python
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['avg_score']=df['total_score']/3
```

```python
sbn.histplot(x=df['avg_score'],kde=True)
```

```python
sbn.histplot(x=df['avg_score'],kde=True,hue=df['gender'])
```

```python
#from grphs above we can see that the average of female student is better than male, hence they tend to perform better.
```

```python
sbn.barplot(x=df['avg_score'],y=df['parental_level_of_education'],hue=df['gender'])
```

```python
sbn.histplot(data=df,x='avg_score',hue='lunch',kde=True)
plt.show()
sbn.histplot(data=df[df['gender']=='female'],x='avg_score',hue=df['lunch'],kde=True)
plt.show()
sbn.histplot(data=df[df['gender']=='male'],x='avg_score',hue=df['lunch'],kde=True)
plt.show()
```

```python
sbn.histplot(x=df['avg_score'],hue=df['race_ethnicity'])
plt.show()
sbn.histplot(data=df[df['gender']=='female'],x=df['avg_score'],hue=df['race_ethnicity'],kde=True)
plt.show()
sbn.histplot(data=df[df['gender']=='male'],x=df['avg_score'],hue=df['race_ethnicity'],kde=True)
plt.show()
```
